```python
import requests
import yaml

# load github access token from yaml file
with open('secretKey.yml') as f:
    secretKey = yaml.safe_load(f)

# GET /user
# Authorization: Bearer <YOUR-TOKEN>
headers = {'Authorization': 'Bearer ' + secretKey['key']}
url = 'https://api.github.com/user'

r = requests.get(url, headers=headers)

```

```python
# parse that
data = r.json()

print (data)

print(f"GitHub username: {data['login']}")
print(f"Number of followers: {data['followers']}")
print(f"Number of repositories: {data['public_repos']}")
```

```python
#Building software Homework 4:
#Working with APIs

import requests
import pandas as pd

# Define the base URL for the Toronto Open Data API
base_url = "https://ckan0.cf.opendata.inter.prod-toronto.ca"

# Define the URL to access the 'Toronto Neighborhoods' API
url = base_url + "/api/3/action/package_show"

# Set the parameters to retrieve information about the 'neighbourhoods' package
params = {"id": "neighbourhoods"}

# Send a GET request to retrieve information about the 'neighbourhoods' package
response = requests.get(url, params=params)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    parse_info = response.json()
    
    # Extract the resources from the package
    resources = parse_info['result']['resources']
    
    # Find the resource with the desired data format (e.g.JSON)
    for resource in resources:
        if resource['format'].lower() in ['json']:
            data_url = resource['url']
            data_format = resource['format'].lower()

#Load the data using JSON obtained above
if data_format == 'json':
        # If the data format is JSON, load it into a DataFrame
        df = pd.read_json(data_url)
else:
        print("Unsupported data format:", data_format)

# Parse and show the DataFrame
df.head()
```
